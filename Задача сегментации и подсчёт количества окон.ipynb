## Расчет количества окон

**Цель:**
1. Определить количества окон в здании фасада
2. Сегментация фасада и окон

Дан датасет из 1134 файлов. Было решено работать с файлами расширения jpg, следовательно проходиться будем чисто по ним.

```python
# Импортируем нужные библиотеки
import cv2
import numpy as np
import os
```

```python
# Путь к директории с изображениями фасадов зданий
images_path = "C:/zzz/CMP_facade_DB_base/base/"

# Пустой словаь для помещения туда колличества(значение) окон по имени файла(ключ)
num_windows_dict = {}

# Цикл для загрузки изображений
for filename in os.listdir(images_path):
    # Проверяем, является ли файл изображением
    if filename.endswith(".jpg"):
        # Чтение изображения
        image = cv2.imread(images_path+filename)
        
        # Конвертирование изображения в оттенки серого
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Применение бинаризации, чтобы преобразовать изображение в двоичное
        _, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
        
        # Применение морфологического преобразования для удаления шума и заполнения мелких дырок
        kernel = np.ones((16,16), np.uint8)
        morphed = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
        
        # Выделение контуров окон на фасаде здания
        contours, _ = cv2.findContours(morphed, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        # Отображение найденных контуров окон на исходном изображении
        image_with_contours = image.copy()
        cv2.drawContours(image_with_contours, contours, -1, (0, 255, 0), 2)
        
        # Подсчет количества окон на фасаде здания
        window_count = 0
        
        for contour in contours:
            # Определение прямоугольника, описывающего контур окна
            x, y, w, h = cv2.boundingRect(contour)
            # Проверка соотношения сторон прямоугольника
            if 0.7 < w/h < 2:
                # Увеличение счетчика окон
                window_count += 1
                
        # Добавление числа окон в словарь
        num_windows_dict[filename] = window_count   
```

Логика нашего подсчёта заключалась в поиске квадратных и прямоугольных контуров. Посмотрев фотографии здания, увидели, что соотношения сторон прямоугольника могут быть различными и постарались учесть их все. Результаты вычислений были занесены в словарь.

## Сегментация фасада и окон

```python
# Путь к папке для сохранения размеченных изображений
labeled_images_path_facade = 'C:/zzz/CMP_facade_DB_base/mask/facade/'
labeled_images_path_window = 'C:/zzz/CMP_facade_DB_base/mask/window/'


# Создаем пустой список для хранения изображений
images = []

# Цикл для загрузки изображений
for filename in os.listdir(images_path):
    # Проверяем, является ли файл изображением
    if filename.endswith(".jpg"):
        # Чтение изображения
        img = cv2.imread(images_path+filename)
        
        # Конвертирование изображения в черно-белый формат
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Применение фильтра Canny для выделения контуров объектов
        edges = cv2.Canny(gray, 100, 200)
        
        # Применение морфологических операций для улучшения контуров объектов
        kernel = np.ones((5,5),np.uint8)
        dilation = cv2.dilate(edges,kernel,iterations = 1)
        closing = cv2.morphologyEx(dilation, cv2.MORPH_CLOSE, kernel)
        
        # Выделение фасада
        facade_mask = np.zeros_like(gray)
        facade_mask[200:800, 50:900] = 255
        
        facade = cv2.bitwise_and(closing, facade_mask)
        
        # Выделение окна
        window_mask = np.zeros_like(gray)
        window_mask[600:700, 150:250] = 255
        window = cv2.bitwise_and(closing, window_mask)

        
        # Вывод результатов на экран
        cv2.imshow('Original Image', img)
        cv2.imshow('Canny Edges', edges)
        cv2.imshow('Dilation', dilation)
        cv2.imshow('Morphological Closing', closing)
        cv2.imshow('Facade Segmentation', facade)
        cv2.imshow('Window Segmentation', window)
        
        # Режим ожидания в ms. Если требуется долго рассматривать, то заменить на (0)
        cv2.waitKey(100)
        
        # Сохранение результатов сегментации фасада и окна
        cv2.imwrite(labeled_images_path_facade+filename, facade)
        cv2.imwrite(labeled_images_path_window+filename, window)
        
        # закрыть все окна
        cv2.destroyAllWindows()
```

Алгоритм подсчёта количества окон основан на выделении контуров определённой длины и определенного соотношения в сторон прямоугольника. Алгоритм сегментации фасада и окон так же основан на выделении контуров объектов на изображении с помощью фильтра Canny, применении морфологических операций для улучшения контуров, и использовании бинарных масок для выделения фасада и окон на изображении.**
